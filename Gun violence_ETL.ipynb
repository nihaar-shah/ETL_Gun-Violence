```python
import pandas as pd
from sqlalchemy import create_engine
import psycopg2
```

### Storing, cleaning and renaming columns of accidental deaths csv

```python
#Store accidental deaths csv into a dataframe
csv_file_1=pd.read_csv('Z:\ETL Project\Accidental deaths.csv')
csv_file_1.head()
```

```python
#Clean accidental deaths csv
new_accidental_deaths=csv_file_1[['Incident ID','Incident Date','State','City Or County','# Killed','# Injured']].copy()
new_accidental_deaths.head()
```

```python
#Rename columns of accidental deaths csv
final_accidental_deaths=new_accidental_deaths.rename(columns={'Incident ID':'incident_id','Incident Date':'incident_date','State':'state',
                                                              'City Or County':'city_or_county','# Killed':'num_killed','# Injured':'num_injured'})
final_accidental_deaths.head()
```

### Storing, cleaning and renaming columns of mass shooting csv

```python
#Store mass shooting csv into a dataframe
csv_file_2=pd.read_csv('Z:\ETL Project\Mass shootings.csv')
csv_file_2.head()
```

```python
#Clean mass shooting csv
new_mass_shooting=csv_file_2[['Incident ID','Incident Date','State','City Or County','# Killed','# Injured']].copy()
new_mass_shooting.head()
```

```python
#Rename columns of mass shooting csv
final_mass_shooting=new_mass_shooting.rename(columns={'Incident ID':'incident_id','Incident Date':'incident_date','State':'state',
                                                              'City Or County':'city_or_county','# Killed':'num_killed','# Injured':'num_injured'})
final_mass_shooting.head()
```

### Storing, cleaning and renaming columns of accidental injuries csv

```python
#Store accidental injuries csv into a dataframe
csv_file_3=pd.read_csv('Z:\ETL Project\Accidental injuries.csv')
csv_file_3.head()
```

```python
#Clean accidental injuries csv
new_accidental_injuries=csv_file_3[['Incident ID','Incident Date','State','City Or County','# Killed','# Injured']].copy()
new_accidental_injuries.head()
```

```python
#Rename columns of accidental injuries csv
final_accidental_injuries=new_accidental_injuries.rename(columns={'Incident ID':'incident_id','Incident Date':'incident_date','State':'state',
                                                              'City Or County':'city_or_county','# Killed':'num_killed','# Injured':'num_injured'})
final_accidental_injuries.head()
```

### Create a connection to the database and load accidental deaths, accidental_injuries and mass_shooting dataframe to the database 

```python
#Create an engine and connection to the postgres database
engine = create_engine('postgres://postgres:PASSWORD@localhost:5432/ETL_Gun_Violence_Project')
conn = engine.connect()
```

```python
#Use pandas to load accidental deaths dataframe into database 
final_accidental_deaths.to_sql('accidental_deaths',con=engine,if_exists='append',index=False )
```

```python
#Check to see if data has been loaded
pd.read_sql_query('select * from accidental_deaths limit 15',con=engine)
```

```python
#Use pandas to load accidental injuries dataframe into database 
final_accidental_injuries.to_sql('accidental_injuries',con=engine,if_exists='append',index=False)
```

```python
#Check to see if data has been loaded
pd.read_sql_query('select * from accidental_injuries limit 15',con=engine)
```

```python
#Use pandas to load mass shooting dataframe into database 
final_mass_shooting.to_sql('mass_shooting',con=engine,if_exists='append',index=False)
```

### Queries to answer project questions

```python
#Which state has the highest number of mass shooting killed incidents

#Removing unnecessary columns 
Mass_shooting_killed_incidents=final_mass_shooting.drop(['incident_id','incident_date','city_or_county','num_injured'],axis=1)
Mass_shooting_killed_incidents

#Total number of mass shooting  
Mass_shooting_killed_incidents=Mass_shooting_killed_incidents.groupby(['state']).sum().sort_values(['num_killed'],ascending=False)
Mass_shooting_killed_incidents
```

```python
#Which state has the highest number of mass shooting injured incidents

#Removing unnecessary columns 
Mass_shooting_injured_incidents=final_mass_shooting.drop(['incident_id','incident_date','city_or_county','num_killed'],axis=1)
Mass_shooting_injured_incidents

#Total number of mass shooting  
Mass_shooting_injured_incidents=Mass_shooting_injured_incidents.groupby(['state']).sum().sort_values(['num_injured'],ascending=False)
Mass_shooting_injured_incidents
```

```python
#Which state has the highest number of killed in accidental deaths csv

#Removing unnecessary columns 
Accidental_deaths_killed_incidents=final_accidental_deaths.drop(['incident_id','incident_date','city_or_county','num_injured'],axis=1)
Accidental_deaths_killed_incidents


#Total number of accidental deaths 
Accidental_deaths_killed_incidents=Accidental_deaths_killed_incidents.groupby(['state']).sum().sort_values(['num_killed'],ascending=False)
Accidental_deaths_killed_incidents
```

```python
#Which state has the highest number of injured in accidental deaths csv

#Removing unnecessary columns 
Accidental_deaths_injured_incidents=final_accidental_deaths.drop(['incident_id','incident_date','city_or_county','num_killed'],axis=1)
Accidental_deaths_injured_incidents


#Total number of accidental deaths 
Accidental_deaths_injured_incidents=Accidental_deaths_injured_incidents.groupby(['state']).sum().sort_values(['num_injured'],ascending=False)
Accidental_deaths_injured_incidents
```

```python
#Which state has the highest number of killed in accidental injuries csv

#Removing unnecessary columns 
Accidental_injuries_killed_incidents=final_accidental_injuries.drop(['incident_id','incident_date','city_or_county','num_injured'],axis=1)
Accidental_injuries_killed_incidents

#Total number of accidental deaths 
Accidental_injuries_killed_incidents=Accidental_injuries_killed_incidents.groupby(['state']).sum().sort_values(['num_killed'],ascending=False)
Accidental_injuries_killed_incidents
```

```python
#Which state has the highest number of injuries in accidental injuries csv

#Removing unnecessary columns 
Accidental_injuries_injured_incidents=final_accidental_injuries.drop(['incident_id','incident_date','city_or_county','num_killed'],axis=1)
Accidental_injuries_injured_incidents

#Total number of accidental deaths 
Accidental_injuries_injured_incidents=Accidental_injuries_injured_incidents.groupby(['state']).sum().sort_values(['num_injured'],ascending=False)
Accidental_injuries_injured_incidents
```

### Merge queried datasets into table to be pushed into databse 

```python
mass_shooting_merge=pd.merge(Mass_shooting_injured_incidents,Mass_shooting_killed_incidents,on='state')
mass_shooting_merge
```

```python
accidental_death_merge=pd.merge(Accidental_deaths_killed_incidents,Accidental_deaths_injured_incidents,on='state')
accidental_death_merge
```

```python
accidental_injury_merge=pd.merge(Accidental_injuries_killed_incidents,Accidental_injuries_injured_incidents,on='state')
accidental_injury_merge
```
